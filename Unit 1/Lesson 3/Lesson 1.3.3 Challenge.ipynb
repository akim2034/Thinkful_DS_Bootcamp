```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("ted_main.csv")

df.columns
```

```python
sns.distplot(df['views']);
```

```python
sns.distplot(df['duration']);
```

```python
sns.distplot(df['comments']);
```

```python
sns.jointplot(x='views', y="duration", data=df);
```

```python
sns.jointplot(x="views", y="comments", data=df);
```

```python
sns.jointplot(x="duration", y="comments", data=df);
```

```python
sns.stripplot(x="views", y="languages", data=df);
```

Distributions are generally the same. They fall under a bell curve with a skew to the right. The relationship between comments and views (r=.53) is stronger than duration and views (r=.05). 

For views and languages, it is not clear what type of relationship the two variables have.
